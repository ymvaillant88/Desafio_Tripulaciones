```python
# Load data from json file
def load_data(file):
    '''
    Función para cargar datos desde un archivo json 
    Args:
        file (str): nombre del archivo json
    Returns:
        data (dict): diccionario con los datos del archivo
    '''
    import json
    with open(file) as f:
        data = json.load(f)
    return data
```
