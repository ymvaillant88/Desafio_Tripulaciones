```python

def encriptar_password(password):
    '''
    Función que encripta una contraseña utilizando el algoritmo SHA-256
    Argumentos:
        - password: contraseña a encriptar
    Devuelve:
        - password_encriptado: contraseña encriptada
    '''
    # Convertir la contraseña a bytes
    password_bytes = password.encode('utf-8')

    # Crear un objeto hash utilizando el algoritmo SHA-256
    hash_object = hashlib.sha256()

    # Actualizar el objeto hash con la contraseña en bytes
    hash_object.update(password_bytes)

    # Obtener la representación hexadecimal del hash
    password_encriptado = hash_object.hexdigest()

    # Devolver el password encriptado
    return password_encriptado

# Encriptar una contraseña
contrasena = input("Ingresa tu contraseña: ")
contrasena_encriptada = encriptar_password(contrasena)
print("La contraseña encriptada es:", contrasena_encriptada)
```

```python
def generateStudents(gender, language, course, id_course):
    '''
    Función que genera un estudiante aleatorio con los datos proporcionados:
    Argumentos:
        gender: género del estudiante
        language: idioma del estudiante
        course: curso del estudiante
    Devuelve:
        student: diccionario con los datos del estudiante
    '''
    faker = Faker(language["locale"])
    random_skills=[]
    random_skills_got=[]
    random_skills_need=[]
    if gender == "male":
        name = faker.name_male()
    else:
        name = faker.name_female()

    names = name.lower().split()
    email = f"{names[0]}.{names[1]}@edem.es"
    birth_date = faker.date_of_birth(minimum_age=18, maximum_age=40)
    # Genera intereses
    random_numberI = random.choice([1, 2, 3, 4, 5])
    random_interests = random.sample(
        ids_interests, k=min(random_numberI, len(ids_interests)))
    # Genera Lenguajes
    random_numberL = random.choice([1, 2, 3, 4])
    remaining_languages = [
        lang for lang in languagesNames if lang != language["name"]]
    random_languages = random.sample(
        remaining_languages, k=min(random_numberL, len(remaining_languages)))
    selected_languages = [language["name"]] + random_languages
    # Generar país de acuerdo al idioma escogido
    country = language["country"]

    
    # Genera conocimientos
    #random_conocimientos = random.sample(
     #   ids_profesional, k=min(random_numberI, len(ids_profesional)))
    
    random_conocimientos, quieroAprender = generate_skills(course)

    # Generar random situacion laboral de acuerdo al curso escogido: si es de un curso de grado, será estudiante, si es de otro curso, será cualquiera
    if "Grado" in course:
        situationLaboral = "Estudiante"
    else:
        situationLaboral = random.choice([situations])


    

    # generate id from mongodb
    new_id = ObjectId()

    # Convertir el ObjectId a una cadena
    str_id = str(new_id)
    return {
        # "_id": str_id,
        "name": name,
        "email": email,
        "password": "728e6cca58a5b904a5a9c6bac40bfeae8733909ee7d7191d042a6e8dd241e136",
        #"type": "student",
        "languages": selected_languages,
        "birth_date": birth_date,
        "country": country,
        "name_course": course,
        "course": id_course,
        "situationLaboral": situationLaboral,
        "role": "user",
        "dateCreated": datetime.datetime.utcnow().isoformat(),
        "gender": gender,
        "eventsConfi": [],
        "imageUrl":faker.image_url(),
        "firstOnBoard": True,
        "conocimientos": random_conocimientos,
        "quieroAprender": quieroAprender,
        "interests": random_interests
        
        
    }
```

## Generación de datos aleatorios para los estudiantes 

```python
# Genera estudiantes de grado
students = []
for career in studentsLevels:
    for course in career["courses"]:
        for i in range(1, (course["male"]+1)):
            language = random.choice(languages)
            students.append(generateStudents("male", language, course["name"], course["id_course"]))
        for i in range(1, (course["female"]+1)):
            language = random.choice(languages)
            students.append(generateStudents(
                "female", language, course["name"], course["id_course"]))
```

```python
# Genera estudiantes de bootcamp
for bootcamp in studentBootcamps:
    for group in bootcamp["groups"]:
        for i in range(1, (group["male"]+1)):
            language = random.choice(languages)
            students.append(generateStudents("male", language,
                            f"{group['name']}", group["id_course"]))
        for i in range(1, (group["female"]+1)):
            language = random.choice(languages)
            students.append(generateStudents("female", language,
                            f"{group['name']}", group["id_course"]))
```

```python
# Genera estudiantes de máster
for master in studentsMaster:
    for group in master["groups"]:
        for i in range(1, (group["male"]+1)):
            language = random.choice(languages)
            students.append(generateStudents("male", language,
                            f"{group['name']}", group["id_course"]))
        for i in range(1, (group["female"]+1)):
            language = random.choice(languages)
            students.append(generateStudents("female", language,
                            f"{group['name']}", group["id_course"]))
```

```python
# Genera dataframe de estudiantes
df_students = pd.DataFrame(students)
df_students.head()
```
