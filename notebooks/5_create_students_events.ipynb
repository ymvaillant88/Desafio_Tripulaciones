## Insertar datos en la base de datos de MongoDB con Python 

```python
# Load data from json file
def load_data(file):
    '''
    Función para cargar datos desde un archivo json 
    Args:
        file (str): nombre del archivo json
    Returns:
        data (dict): diccionario con los datos del archivo
    '''
    import json
    with open(file) as f:
        data = json.load(f)
    return data
```

## Insertar eventos

```python
interests = ["Caminar", "Correr", "Cine", "Libros", "Revistas", "Fútbol", "Baloncesto", "Tenis", "Natación", "Senderismo",
             "Videojuegos", "Ajedrez", "Voluntariados", "Conciertos", "Cultura", "Restaurantes", "Bailar", "Gimnasio",
             "Yoga", "Viajar", "Quedar con amigos", "Actividades al aire libre", "Acampar", "Visitar museos", "Hacer picnic",
             "Exposiciones", "Fotografía", "Meditación", "Spa", "Turismo"]

# GEt 10 random interests
random_interests = random.sample(interests, 20)
random_interests
```

```python
# Obtener todos los tags donde profesional sea False
filter = {"profesional": False}

tagsDocs = db["tags"].find(filter)
tags = []
for doc in tagsDocs:
    tags.append(doc)

tags
```

```python
# Listado con nombres de localizaciónes aleatorias de la ciudad de Valencia 
locations = [
    "Valencia",
    "Alboraya",
    "Alcàsser",
    "Aldaia",
    "Alfafar",
    "Almàssera"]
```

```python
# create events from random_interests
events = []
for interest in random_interests:
    now = datetime.datetime.now()
    user1 = random.choice(students)
    user2 = random.choice(students)
    # Generar fecha incio y fin del evento diferente a la actual (guardada en la variable now) y que la fecha de inicio sea menor a la fecha de fin y con 1 hora de duración
    start = now + datetime.timedelta(days=random.randint(1, 30)) + datetime.timedelta(hours=random.randint(1, 23)) + datetime.timedelta(minutes=random.randint(1, 59)) + datetime.timedelta(seconds=random.randint(1, 59))
    end = start + datetime.timedelta(hours=1) + datetime.timedelta(minutes=random.randint(1, 59)) + datetime.timedelta(seconds=random.randint(1, 59)) + datetime.timedelta(days=random.randint(1, 30))
    # Generar localizacion del evento
    location = random.choice(locations)
    # Generar el id tags del evento (1 o 2) aleatorios de la lista de tags
    tag = random.choice(tags)
    tag = str(tag["_id"])

    events.append({
        "name": "Evento de " + interest,
        "description": "Evento de " + interest,
        "createdBy": user1["_id"],
        "date": now,
        "startDate": start,
        "endDate": end,
        "localization": location,
        "attendees": [user2["_id"]],
        #"interest": interest,
        "imageUrl": "https://picsum.photos/200/300",
        "tags": [tag],
        "oficial": False


    })

events
```

```python
# Convertir a dataframe events
df_events = pd.DataFrame(events)
df_events
```
